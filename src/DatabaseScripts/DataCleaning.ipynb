```python
import json
import csv
headers = True
review_file = './data/yelp-dataset/data_CSV/data_compressed/Restaurant_business.json'
review_output='./data/yelp-dataset/data_CSV/data_compressed/Restaurant_business_check.csv'




with open(review_file,encoding="utf-8" ) as jsonf, open(review_output,"w",encoding="utf-8") as csvf:
    for line in jsonf:
        data = json.loads(line)
        print(data)
        if headers:
            keys =[]
            for k,v in data.items():
                keys.append(k)
                writer = csv.DictWriter(csvf,fieldnames=keys,lineterminator = '\n')
            writer.writeheader()
            headers=False
        else:        
            writer.writerow(data)
                
        

```

```python
user_df_curr=user_df[['user_id', 'name' ,'review_count', 'yelping_since', 'useful', 'fans', 'elite', 'average_stars']]
```

```python
import re
import pandas as pd
import numpy as np




user_df_curr=user_df_curr.dropna(how='any')

user_df_curr['Impact_Score']=0.0
user_df_curr['Normalized_Impact_Score']=0.0


newIndex=0
totalValue = 34
for index, row in user_df_curr.iterrows():
    if row['review_count']!=0 and len(re.findall("[a-zA-Z]", str(row['review_count'])))==0 and len(re.findall("[a-zA-Z]", str(row['useful'])))==0 and len(re.findall("[a-zA-Z]", str(row['fans'])))==0 :
        
        value=0.0
        value=value+((row['useful']/row['review_count'])*9)+(len(row['elite'].split(','))*10)+(8*row['average_stars'])+(7*row['fans'])
        user_df_curr.at[index,'Impact_Score']=value/totalValue
    else:
        user_df_curr.drop(index, inplace=True)
            
       

```
